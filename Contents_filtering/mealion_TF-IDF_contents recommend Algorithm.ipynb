```python
review_restaurant = pd.merge(restaurants,reviews, left_on='id',right_on='restaurantId', how='outer' )
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer


# ngram_range=(1, 2) : ë‹¨ì–´ë¥¼ 1ê°œ í˜¹ì€ 2ê°œ ì—°ì†ìœ¼ë¡œ ë³´ê² ë‹¤
tfidf_vec = TfidfVectorizer(ngram_range=(1,2))
tfidf_matrix = tfidf_vec.fit_transform(review_restaurant['content'])
```

```python
tfidf_matrix 
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# 4803ê°œì˜ ì˜í™”ë‘ 4803ê°œì˜ ì˜í™”ë¼ë¦¬ ìœ ì‚¬ë„ë¥¼ êµ¬í•˜ê² ë‹¤!
# ìì‹ ê³¼ì˜ ìœ ì‚¬ë„ëŠ” 1
genres_similarity = cosine_similarity(tfidf_matrix, tfidf_matrix)
print(genres_similarity)
```

```python
# ìœ ì‚¬ë„ ê°’ì´ ë†’ì€ ì˜í™”ì˜ ì œëª©
# ìœ ì‚¬ë„ ê°’ì´ ë†’ì€ ìˆœìœ¼ë¡œ ì¸ë±ìŠ¤ ê°’ì„ ë½‘ì•„ë‚¸ë‹¤
similar_index = np.argsort(-genres_similarity)
print(similar_index)
#np.argsort(-genres_similarity)ë¥¼ ìˆ˜í–‰í•˜ë©´ genres_similarityë¥¼ ìœ ì‚¬ë„ê°€ ë†’ì€ ìˆœì„œëŒ€ë¡œ ì •ë ¬í•œ ì¸ë±ìŠ¤ numpy ë°°ì—´ì„ ë¦¬í„´í•œë‹¤. 
#ì¦‰, ê° ì˜í™”ì— í•´ë‹¹í•˜ëŠ” í–‰ë§ˆë‹¤ ìì‹ ê³¼ ìœ ì‚¬í•œ ì˜í™”ë¥¼ ìˆœì„œëŒ€ë¡œ ê°€ì§€ê³  ìˆëŠ” ë°°ì—´ì´ ìƒì„±ë˜ëŠ” ê²ƒì´ë‹¤.
```

```python
reviews = pd.read_csv('C:/Users/kimjunseok/Downloads/mealion_reviews.csv')
```

```python
reviews.rename(columns = {'id' : 'userid'}, inplace = True)

```

```python
reviews.head()
```

```python
reviews.info()
```

```python
review_restaurant = pd.merge(restaurants,reviews, left_on='id',right_on='restaurantId', how='outer' )
```

```python
review_restaurant = review_restaurant.loc[review_restaurant.content.notna(),: ]
```

```python
review_restaurant.reset_index(drop=True, inplace=True)
```

```python
review_restaurant
```

```python
review_restaurant['content']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer


# ngram_range=(1, 2) : 단어를 1개 혹은 2개 연속으로 보겠다
tfidf_vec = TfidfVectorizer(ngram_range=(1,2))
tfidf_matrix = tfidf_vec.fit_transform(review_restaurant['content'])
```

```python
tfidf_matrix 
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# 4803개의 영화랑 4803개의 영화끼리 유사도를 구하겠다!
# 자신과의 유사도는 1
genres_similarity = cosine_similarity(tfidf_matrix, tfidf_matrix)
print(genres_similarity)
```

```python
# 유사도 값이 높은 영화의 제목
# 유사도 값이 높은 순으로 인덱스 값을 뽑아낸다
similar_index = np.argsort(-genres_similarity)
print(similar_index)
#np.argsort(-genres_similarity)를 수행하면 genres_similarity를 유사도가 높은 순서대로 정렬한 인덱스 numpy 배열을 리턴한다. 
#즉, 각 영화에 해당하는 행마다 자신과 유사한 영화를 순서대로 가지고 있는 배열이 생성되는 것이다.
```

```python
review_restaurant.name
```

```python
input_restaurant = input()

restaurant_index = review_restaurant[review_restaurant['name']==input_restaurant].index.values
similar_restaurants = similar_index[restaurant_index, :10]

# 인덱스로 사용하기 용이하도록 1차원으로 변경
similar_restaurant_index = similar_restaurants.reshape(-1)

display(review_restaurant.iloc[similar_restaurant_index])
```

우선 추천 순위를 TF-IDF, 리뷰가 부족한 곳에는 그 다음 순위로 Countervectorizer도 식당을 추천해준다. 그럼 데이터가 1000개에서 2000개로 늘어난다.... 
1순위) TF-IDF : 리뷰를 분석해 유사한 추천 맛집을 알려준다- 더 사용자 취향에 어울리는 세심하게 맛집을 추천해줄 수 있다.
2순위) CounterVectorizer: 같은 카테고리(한식, 일식, 양식...) 순으로, 평점 수, 평점 평균 순으로 추천해준다. - 덜 세심하지만, 그럴 듯하게 사용자에게 맞는 맛집을 추천해준다.
3순위)
-권장: 리뷰와 평점 데이터를 많이 확보하다. 
많이 확보하면 
1만개 이상으로 (85만개 그정도로 필요 없으니깐, 인기있는집, 알게 모르게 많이가는 집, 새로 생긴 음식점 위주로 1~5만개만 뽑아도 된다) -> 리뷰와 평점을 어떻게 얻지? 
1. 공공데이터나 사적 데이터 얻기

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
count_vector = CountVectorizer(ngram_range=(1,3))
```

```python
c_vector_category = count_vector.fit_transform(small_restaurant_df ['category'])
```

```python
c_vector_category.shape
```

```python
category_c_sim = cosine_similarity(c_vector_category, c_vector_category).argsort()[:, ::-1]
```

```python
category_c_sim.shape
```

```python
def get_recommand_restaurant_list(restaurant_df, restaurant_title, top=30 ):
    target_restaurant_index = restaurant_df[restaurant_df['restaurant_title'] == restaurant_title].index.values
    sim_index = category_c_sim[target_restaurant_index, :top].reshape(-1)
    sim_index = sim_index[sim_index != target_restaurant_index]
    result = restaurant_df.iloc[sim_index].sort_values('score', ascending=False)[:10]
    return result

    
```

```python
get_recommand_restaurant_list(small_restaurant_df, restaurant_title='중화루')
```
